## 1. Visualization Technique  &nbsp;·&nbsp; *the five charts and why they work together*

A dashboard is not five charts stacked on a page — it is five *lenses on one
question*. Our question: **how has the World Cup changed over nearly 100 years?**
Each chart answers one facet, and the interactivity is what turns the collection
into a single instrument.

### 1.1 Line chart — *change over time*
A line chart is the right tool when the x-axis is continuous and ordered, and the
story is a **trajectory**. We use it for goals-per-match across editions. The eye
naturally reads slope as "rising" or "falling," which is exactly the scoring
narrative. We overlay two extra layers: gold **markers for rule changes** (VAR,
field expansions, the back-pass rule) so trends can be tied to causes, and gold
**stars** marking editions won by a selected confederation.

**Use it when:** you have an ordered sequence and want to show direction and
inflection points. **Avoid it when:** the x-axis is categorical and unordered — a
bar chart is clearer there.

### 1.2 Bar chart — *magnitude comparison across categories*
Bars compare discrete quantities; length is the most accurately perceived visual
encoding humans have. We use bars for cards-per-match by edition (data exists only
from 1970, when cards were introduced — a good lesson in *honest gaps*). Colour
doubles the encoding (darker = more cards) so intensity reads at a glance.

**Use it when:** comparing a value across a modest number of discrete categories.
**Avoid it when:** you have dozens of bars (becomes a picket fence) or a true
time-trend where a line reads better.

### 1.3 Scatter plot — *relationship and outliers*
Scatter plots reveal how two variables relate and surface outliers. We plot each
host nation's **performance lift** (points-per-match at home minus away) against
year, size the points by matches played as host, and colour them by whether the
host over- or under-performed. A dashed zero-line makes "better at home" instantly
legible.

**Use it when:** you want to show correlation, spread, or exceptions.
**Avoid it when:** one axis is a small ordered category — a line or bar is tidier.

### 1.4 Treemap — *part-to-whole across a hierarchy*
Treemaps pack a hierarchy into nested rectangles sized by value. Ours nests
**confederation → nation**, sized by finals reached and coloured by titles won. It
communicates in one glance that only two confederations have *ever* reached a
final — a finding a table would bury.

**Use it when:** you have hierarchical part-to-whole data and care about relative
size. **Avoid it when:** precise value comparison matters (areas are read less
accurately than lengths) or the hierarchy is deep.

### 1.5 Choropleth map — *geography*
When the dimension is *where*, put it on a map. We shade host nations by their
performance lift on a diverging red-green scale centred at zero. Geography that
would be invisible in a table (the clustering of hosts in Europe and the Americas)
becomes obvious.

**Use it when:** the variable is inherently spatial. **Avoid it when:** area size
distorts the message (large sparse countries dominate) — consider a cartogram.

### 1.6 How they complement each other — and the role of interactivity
Each chart is weak alone. Together they triangulate the same century from four
angles — time, magnitude, relationship, geography — so a reader can move from
"scoring fell" (line) to "discipline rose" (bar) to "hosts still overperform"
(scatter/map) without changing pages.

The **dashboard-specific** consideration is *linked interactivity*. A single
year-range slider filters all five charts and four KPIs **simultaneously**. This
matters because it lets the reader ask their own questions ("what did the
post-1994 era look like?") instead of consuming a fixed answer. The design rule
we follow: **one primary control that touches everything**, plus a lightweight
secondary filter (confederation) that reshapes only the charts where it is
meaningful. Too many independent controls fragment the story; one shared control
keeps every view in sync.

## 2. Visualization Library  &nbsp;·&nbsp; *why Plotly Dash*

### 2.1 What it is, who made it, and how to install it
**Plotly** is an open-source (MIT-licensed) graphing library created by the
Canadian company Plotly Inc.; it renders charts to interactive JavaScript
(D3/WebGL) from a Python API. **Dash** — also by Plotly, also open source — is the
web-application framework layered on top: it turns Plotly figures into a reactive
dashboard without your writing any HTML, CSS, or JavaScript by hand.

```bash
pip install dash plotly pandas
```

We deliberately avoid **Matplotlib, Seaborn, and Altair** for this assignment, as
those (or close cousins) are already covered elsewhere. Plotly Dash brings
something genuinely different: charts that are interactive *by default* and a
first-class app framework.

### 2.2 Declarative or procedural?
It is **both, on purpose**, and knowing the split is the key to using it well:

- **Plotly Express / graph_objects — declarative.** You describe *what* you want
  (`px.bar(df, x=..., y=..., color=...)`) and Plotly figures out the drawing.
- **Dash callbacks — procedural / reactive.** You write functions that say
  *when input X changes, recompute outputs Y*. This is the imperative glue that
  makes charts respond to controls.

Contrast with Altair, which is purely declarative: elegant for static specs, but
Dash's procedural callback layer is what unlocks true multi-chart cross-filtering.

### 2.3 Jupyter integration
Plotly figures render inline in Jupyter with `fig.show()` — which is exactly how
we prototype each chart in this notebook before wiring it into the app. Full Dash
apps can also run inline in a notebook (`app.run(jupyter_mode="inline")`), though
for a permanent public URL we deploy the app as a standalone script instead.

### 2.4 Why this framework, and its limitations
**Why we chose it:** it satisfies every requirement in one stack — four-plus
interactive chart types, a single control that cross-filters all of them, and free
public hosting. It is also the closest of the options to an industry-standard BI
tool, so the skills transfer.

**Limitations to be honest about:**
- **Stateful server.** A live Dash app needs a running Python process, so
  "publish" means deploy to a host (Hugging Face Spaces / Render), not just export
  an HTML file. We handle this with the included `Dockerfile`.
- **Callback verbosity.** Complex dashboards accumulate many `Input`/`Output`
  wirings; we keep ours to a *single* callback for clarity.
- **Not for print.** Output is inherently web/interactive; for a static PDF figure,
  a declarative library is lighter.

## 3. Demonstration  &nbsp;·&nbsp; *building the dashboard, step by step*

### 3.1 The dataset and how it was cleaned
The data lives in `data/` as six CSVs derived from the **Fjelstul World Cup
Database** (1930–2022, CC-BY-SA 4.0) plus a **partial 2026 layer** (through the
quarter-finals). The file that drives most charts is `wc_editions.csv` — one row
per tournament.

**Cleaning steps already applied (and worth replicating on raw data):**
1. **One row per edition**, with pre-computed rates (`goals_per_match`,
   `cards_per_match`) so charts don't recompute on every callback.
2. **Honest gaps kept as `NaN`.** Cards did not exist before 1970, so those cells
   are empty rather than zero — we `.dropna()` per chart instead of imputing.
3. **Partial edition flagged.** 2026 has `is_partial_edition = 1` and no card data.
4. **Milestones tagged.** `has_rule_note = 1` marks editions with a rule change,
   which become annotations on the line chart.
5. **Co-hosts split** in `wc_host_performance.csv` (2002 → South Korea + Japan;
   2026 → USA + Canada + Mexico) so each maps to one country on the choropleth.

````python
import pandas as pd

editions = pd.read_csv("data/wc_editions.csv")
hosts    = pd.read_csv("data/wc_host_performance.csv")

print(editions.shape, "editions;", hosts.shape, "host rows")
editions[["year", "host", "goals_per_match", "cards_per_match", "champion"]].head()
````

### 3.2 A basic component first
Always get one static chart right before adding any interactivity. Here is the
goals-per-match line on its own — the atom the dashboard is built from.

````python
import plotly.graph_objects as go

d = editions.dropna(subset=["goals_per_match"]).sort_values("year")

fig = go.Figure(go.Scatter(
    x=d["year"], y=d["goals_per_match"],
    mode="lines+markers", line=dict(color="#1a7a4c", width=3),
))
fig.update_layout(title="Goals per match, 1930-2026",
                  xaxis_title="Edition", yaxis_title="Goals / match",
                  paper_bgcolor="white", plot_bgcolor="white")
fig.show()
````

Confirm each of the other four (bar, scatter, treemap, choropleth) renders on its
own the same way. The full, themed builders live in `app.py` as `fig_goals_line`,
`fig_cards_bar`, `fig_host_scatter`, `fig_confed_treemap`, and `fig_host_map`. Here
is the treemap builder as a second example, since hierarchy setup is the least
obvious:

````python
import plotly.express as px

rows = []
for _, r in editions.iterrows():
    rows.append((r["champion_confed"], r["champion"], 1))
    rows.append((r["runner_up_confed"], r["runner_up"], 0))
g = (pd.DataFrame(rows, columns=["confed", "nation", "is_title"])
       .groupby(["confed", "nation"], as_index=False)
       .agg(finals=("nation", "size"), titles=("is_title", "sum")))

fig = px.treemap(g, path=[px.Constant("All"), "confed", "nation"],
                 values="finals", color="titles",
                 color_continuous_scale=["#cfe4d8", "#1a7a4c", "#0f4d30"])
fig.show()
````

### 3.3 Adding interactivity — the cross-filter callback
This is the heart of the assignment. In Dash, the layout declares *controls* and
*empty chart slots*; a **callback** connects them. The pattern is: the slider's
value is an `Input`, each chart's `figure` is an `Output`, and one function
recomputes them all. Because every chart reads from the same filtered frame, one
slider move updates the entire dashboard at once.

```python
from dash import Dash, dcc, html, Input, Output

app = Dash(__name__)
app.layout = html.Div([
    dcc.RangeSlider(id="year-range", min=1930, max=2026,
                    value=[1930, 2026], step=None, allowCross=False),
    dcc.Graph(id="g-line"),
    dcc.Graph(id="g-bar"),
    # ... three more chart slots ...
])

@app.callback(
    Output("g-line", "figure"), Output("g-bar", "figure"),  # + 3 more
    Input("year-range", "value"),
)
```

### 3.5 Troubleshooting — errors you are likely to hit
| Symptom | Cause | Fix |
|---|---|---|
| `AttributeError: 'Dash' object has no attribute 'run_server'` | Dash 3 renamed the method | Use `app.run(...)` (already used here) |
| Choropleth is blank for a country | Plotly doesn't know the name | Map it: `{"West Germany": "Germany", "England": "United Kingdom"}` |
| Bar chart empty for early years | No card data before 1970 | Expected — `.dropna()` and show a note, don't impute zeros |
| Callback error: outputs ≠ returns | Output count must equal returned values | Count `Output(...)` decorators = items in the `return` tuple |
| Space builds but page won't load | Wrong port | Serve on `7860`; set `app_port: 7860` in the Space README |
